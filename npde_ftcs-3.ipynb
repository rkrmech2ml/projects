```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
L = np.pi     # length
T = 1.0     # total 
N = 10    # number of grid points
M = 100    # number of time steps
kappa = 1 # diffusion coefficient 
dx = L/N    # grid spacing
dt = T/M    # time step
gamma=(dt/dx**2)
print(gamma)

# Create the grid
x = np.linspace(0, L, N+1)
print(len(x))
t = np.linspace(0, T, M+1)

# Set the initial condition
u = np.zeros((N+1, M+1))
utrue = np.zeros((N+1, M+1))
u[:,0] = np.sin(x)
#print(u)

# Apply the explicit Euler method
for k in range(M):
    for i in range(1, N):
        u[i,k+1] = u[i,k] + kappa*dt/dx**2*(u[i-1,k] - 2*u[i,k] + u[i+1,k])

# Plot the solution
print(u)
fig, ax = plt.subplots()
ax.plot(x, u[:,0], label='t=0')
print('u is',u)
print(len(u))
ax.plot(x, u[:,1], label='t=1')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('u')
ax.set_title('Solution of the Parabolic Equation')
plt.show()
```
